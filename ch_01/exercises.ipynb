## Exercise 4: Generate the data by running this cell
This will give you a list of numbers to work with in the remaining exercises.

```python
import random

random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

## Exercise 5: Calculating statistics and verifying
### mean

```python
import statistics as s
```

```python
m = sum(salaries) / len(salaries)
m == s.mean(salaries)
```

### median

```python
salaries.sort()
mid = (int((len(salaries) / 2)) - 1)

if len(salaries)%2 == 0:
    median = (salaries[mid] + salaries[mid+1]) / 2
else:
    median = salaries[mid]
    
median == s.median(salaries)
```

### mode

```python
from collections import Counter 

c = Counter(salaries)
c.most_common(1)[0][0] == s.mode(salaries)
```

### sample variance
Remember to use Bessel's correction.

```python
x = 0
for salary in salaries:
    x = x + (salary - s.mean(salaries)) ** 2
 
variance = x / (len(salaries)-1)
variance == s.variance(salaries)
```

### sample standard deviation
Remember to use Bessel's correction.

```python
import math
std = math.sqrt(variance)
std == s.stdev(salaries)
```

## Exercise 6: Calculating more statistics
### range

```python
max(salaries) - min(salaries)
```

### coefficient of variation
Make sure to use the sample standard deviation.

```python
s.stdev(salaries)/s.mean(salaries)
```

### interquartile range

```python
help(s.quantiles)
```

```python
s.quantiles(salaries,n=4)[2] - s.quantiles(salaries,n=4)[0]
```

### quartile coefficent of dispersion

```python
(s.quantiles(salaries,n=4)[2] - s.quantiles(salaries,n=4)[0]) / (s.quantiles(salaries,n=4)[2] + s.quantiles(salaries,n=4)[0])
```

## Exercise 7: Scaling data
### min-max scaling

```python
new_salaries = [(salary - min(salaries)) / (max(salaries) - min(salaries)) for salary in salaries]
new_salaries
```

```python
new_salaries2 = []
for salary in salaries:
    x = (salary - min(salaries)) / (max(salaries) - min(salaries))
    new_salaries2.append(x)
    
new_salaries2
```

### standardizing

```python
new_salaries = [(salary - s.mean(salaries)) / s.stdev(salaries) for salary in salaries]
new_salaries
```
